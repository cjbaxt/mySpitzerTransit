```python
from matplotlib.gridspec import GridSpec

def centroidRMS(x,y, title = None, width = 150, plot = True):
    
    fig=plt.figure(figsize=(16, 15))

    gs=GridSpec(5,2,height_ratios=[1, 1,1,1,2]) # 5 rows, 2 columns

    ax1=fig.add_subplot(gs[0,:]) # First row, first column
    ax2=fig.add_subplot(gs[1,:]) # First row, second column
    ax3=fig.add_subplot(gs[2,:]) # First row, third column
    ax4=fig.add_subplot(gs[3,:]) # First row, fourth column
    ax5=fig.add_subplot(gs[4,0]) # Second row, span all columns
    ax6=fig.add_subplot(gs[4,1]) # Second row, span all columns

    # rms test 
    width = 150
    smoothx = convolve(x, Box1DKernel(width), boundary='extend')
    smoothy = convolve(y, Box1DKernel(width), boundary='extend')

    rmsx = np.sqrt(np.mean((x - smoothx)**2))
    rmsy = np.sqrt(np.mean((y - smoothy)**2))
    planet = 'planet'
    channel = 'ch1'
    aor = 'aor'
    ax1.set_title("{} {} {}".format(planet,channel,aor))
    ax1.plot(x)
    ax2.plot(y)
    ax1.plot(smoothx, c = 'C2')
    ax2.plot(smoothy, c = 'C2')

    ax3.plot(x-smoothx, c = 'C2')
    ax4.plot(y-smoothy, c = 'C2')

    ax5.hist(x-smoothx,bins=15 , label = '{:.3f}'.format(rmsx))
    ax6.hist(y-smoothy,bins=15 , label = '{:.3f}'.format(rmsy))
    ax5.set_title('rms = {:.3f}'.format(rmsx))
    ax6.set_title('rms = {:.3f}'.format(rmsy))

    plt.savefig(title)
    
    return rmsx, rmsy


print rmsx,rmsy
```
